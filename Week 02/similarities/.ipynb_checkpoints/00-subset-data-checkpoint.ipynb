# Preparing dataset

We will use the BX-Book-Ratings dataset for this week's notebooks. Since the set is rather large, please subset it and save a copy so that you can use it later. 

Please filter out:

* users that have less than 20 ratings
* books that have less than 20 ratings


```python
import pandas as pd
df_books_ratings = pd.read_csv('../../data/BX-Book-Ratings.csv', sep=';', encoding='latin-1')
df_books_ratings
```

```python
df_user_counts = df_books_ratings["User-ID"].value_counts()
df_users = df_user_counts[df_user_counts >= 20]
users = df_users.index.tolist()

len(users)
```

```python
df_book_counts = df_books_ratings["ISBN"].value_counts()
df_books = df_book_counts[df_book_counts >= 20]
books = df_books.index.tolist()
books
```

```python
df = df_books_ratings[df_books_ratings["User-ID"].isin(users)]
df = df_books_ratings[df_books_ratings["ISBN"].isin(books)]
df
```

```python
df.to_csv('data/BX-Book-Ratings-Subset.csv', index=False, sep=';')
```
